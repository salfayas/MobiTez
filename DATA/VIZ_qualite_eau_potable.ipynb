```python
from pymongo import MongoClient
import pandas as pd
from matplotlib import pyplot as plt
# Connexion à la base de données MongoDB
client = MongoClient('localhost', 27017)

# Sélection de la base de données
db = client.TEZEAU

# Sélection de la collection
collection = db.tbl_qualite_eaupotable

# Exécution d'une requête
result = collection.find({"commune": "VICHY"})



df = pd.DataFrame(result)
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%dT%H:%M:%SZ')
df

```

```python
df.info()
```

```python
df_ph = df.loc[df['param'] == 'PH']
df_ph
```

```python

# Tracé de la courbe de l'évolution du pH
plt.plot(df_ph['date'],df_ph['valeur'])
plt.title("Évolution du pH à Vichy")
plt.xlabel("date")
plt.ylabel("pH")
plt.xticks(df_ph['date'][::1], rotation=45)
plt.text(1.05, 0.5, "Référence qualité :\n\n >=6,5 X <=9", transform=plt.gcf().transFigure)

plt.show()
```

```python
df_nitrate = df.loc[df['param'] == 'Nitrate']
```

```python
# Tracé de la courbe de l'évolution du nitrate
plt.plot(df_nitrate['date'],df_nitrate['valeur'])
plt.title("Évolution du nitrate à Vichy")
plt.xlabel("date")
plt.ylabel("nitrate")
plt.xticks(df_nitrate['date'][::1], rotation=45)
plt.text(1.05, 0.5, "Référence qualité :\n\n 0 - 3.0 : Pas d'impact sur la santé \n\n 3.1 - 10 : impact sur la santé non démontré \n\n >10 : impact sur la santé possible", transform=plt.gcf().transFigure)

plt.show()
```

```python
df_Durete_eau = df.loc[df['param'] == 'Durete_eau']
```

```python
# Tracé de la courbe de l'évolution de la Durete_eau
plt.plot(df_Durete_eau['date'],df_Durete_eau['valeur'])
plt.title("Évolution de la dureté de l'eau à Vichy")
plt.xlabel("date")
plt.ylabel("nitrate")
plt.xticks(df_Durete_eau['date'][::1], rotation=45)
plt.text(1.05, 0.5, "Référence qualité :\n\n 0 - 8 : Eau très douce \n\n 8 - 15 : Eau douce \n\n 15-30 : Eau moyennement dure \n\n >30 : Eau très dure", transform=plt.gcf().transFigure)

plt.show()
```
